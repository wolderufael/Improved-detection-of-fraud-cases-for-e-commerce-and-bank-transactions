# Model Explainability
Model explainability is crucial for understanding, trust, and debugging in machine learning models. You will use SHAP (Shapley Additive exPlanations) and LIME (Local Interpretable Model-agnostic Explanations) to interpret the models you built for fraud detection.


### Adding project path to system path

```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
notebook_dir = os.getcwd()
parent_path=os.path.dirname(notebook_dir)

os.chdir(parent_path)
```

### Import modules

```python
from script.SHAP_LIME import Explainaibility
explainaibility=Explainaibility()
```

### Load the preprocessed fraud data

```python
fraud_data=pd.read_csv("Data/preprocessed_data.csv")
```

### Split the data

```python
X_train, X_test, y_train, y_test=explainaibility.split_data(fraud_data)
```

## Model Explainaibility for LogesticRegression

### Load the model

```python
lr_model=explainaibility.load_model("models/fraud_data/LogisticRegression-28-10-2024-10-27-17-00.pkl")
```

### SHAP parameters

```python
explainer,shap_values=explainaibility.shap_params(lr_model,X_test)
```

### Summary Plot

Provides an overview of the most important features.

```python
explainaibility.summary_plot(shap_values,X_test)
```

## Force Plot

Visualizes the contribution of features for a single prediction.


```python
explainaibility.force_plot(shap_values,explainer,X_test)
```

### Dependence Plot

This shows the relationship between a feature and the model output.

```python
explainaibility.dependece_plot(shap_values,"purchase_value",X_test)
```
