```python
import pickle
with open("models/fraud_data/DecisionTreeClassifier-22-10-2024-22-27-33-00.pkl", 'rb') as f:
    model = pickle.load(f)
```

```python
import lime
from lime.lime_tabular import LimeTabularExplainer
import numpy as np

# Assuming X_train is your training data (as a DataFrame or numpy array) and X_test is the test set
explainer = LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X_train.columns,
    class_names=[0, 1],  # Adjust this based on your class labels
    mode='classification'
)

# Select an instance for explanation
instance_to_explain = X_test.iloc[0]  # Choose the instance from the test set

# Explain the prediction for this instance
explanation = explainer.explain_instance(
    data_row=instance_to_explain,
    predict_fn=model.predict_proba  # Assumes your model has a predict_proba method
)

# Plot the explanation (Feature Importance Plot)
explanation.show_in_notebook(show_table=True)
```
